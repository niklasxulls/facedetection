```python
def download_and_unzip(outputname, url, destpath, rename):
    dest = os.path.join(destpath, outputname)
    os.chdir(destpath)
    gdown.download(url, outputname, quiet=False)

    with zipfile.ZipFile(outputname,"r") as zip_ref:
         zip_ref.extractall(destpath)
        
         rootFolder = zip_ref.filelist[0].filename.rstrip('/')
         curr = os.path.join(destpath, rootFolder)
         newname = curr.replace(rootFolder, rename)

         if rename and not os.path.exists(newname):
            os.rename(curr, newname)

    os.chdir(INSTALLATION_PATH)
    os.remove(dest)
```
